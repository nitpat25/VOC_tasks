### Task 1

```python
import string
s = " s Sea shells built shells s 10. The 10 people are alive."
word = input("Enter the word who's frequency you want to check:")
n = len(s)
s = s.translate(str.maketrans('', '', string.punctuation))
n=len(s)
def freq(s,word):
    i=0
    count=0
    freq=0
    m=len(word)
    while i<n and m<=n:
        for char in word:
          if i<n:
            if s[i]==char:
              i=i+1
              count+=1
            else:
              i=i+1
              count=0
              break
        j=i-m
        k=j-1
        if k>=0:
          if i<(n-1):
              if count==m and s[i]==' ' and s[k]==' ':
                count=0
                freq+=1
          elif i>(n-1):
            if m==n:
              if count==m:
                count=0
                freq+=1
            elif m<n:
              if count==m and s[k]==' ':
                count=0
                freq+=1
          elif i==(n-1):
              if m==n:
                if count==m:
                  count=0
                  freq+=1
              elif m<n:
                if k!=0:
                  if count==m and s[k]==' ' and s[i]==' ':
                    count=0
                    freq+=1
                elif k==0 and s[k]==' ':
                  if count==m:
                    count=0
                    freq+=1
        else:
          if i<(n-1):
            if count==m and s[i]==' ':
                  count=0
                  freq+=1
          elif i>(n-1):
            if m==n:
              if count==m:
                count=0
                freq+=1
            elif m<n:
              if count==m and s[i]==' ':
                count=0
                freq+=1
          elif i==(n-1):
              if m==n:
                if count==m:
                  count=0
                  freq+=1
              elif m<n:
                if count==m and s[i]==' ':
                      count=0
                      freq+=1
    return freq

freq(s,word)
```

### Task 2

```python
class To_do:
    def __init__(self):
        self.list=[]
        
    def is_empty(self):
        if not self.list:
            print("You have no tasks to do!")
        else:
            print("Your to_do list is not empty.")
        
    def add_task(self,task):
        self.task=task
        self.list.append(task)
        
    def remove_task(self):
        self.list.pop()
        
    def add_beg(self,task):
        self.task=task
        self.list.insert(0,self.task)
        
    def remove_beg(self):
        self.list.pop(0)
        
    def add_custom(self,pos,task):
        self.task=task
        self.pos=pos-1
        self.list.insert(self.pos,self.task)
        
    def remove_custom(self,pos):
        self.list.pop(pos-1)
        
    def display(self):
        n = len(self.list)
        for i in range(n):
            print(f"Task {i+1}: {self.list[i]}")
            
            
def main():
    todo_list = To_do()

    while True:
        print("\nMenu:")
        print("1. Check if the list is empty")
        print("2. Add a task")
        print("3. Remove the last task")
        print("4. Add a task to the beginning")
        print("5. Remove the first task")
        print("6. Add a task at a custom position")
        print("7. Remove a task from a custom position")
        print("8. Display the to-do list")
        print("9. Quit")

        choice = input("Enter your choice (1-9): ")

        if choice == "1":
            todo_list.is_empty()
        elif choice == "2":
            task = input("Enter the task to add: ")
            todo_list.add_task(task)
        elif choice == "3":
            todo_list.remove_task()
        elif choice == "4":
            task = input("Enter the task to add at the beginning: ")
            todo_list.add_beg(task)
        elif choice == "5":
            todo_list.remove_beg()
        elif choice == "6":
            task = input("Enter the task to add: ")
            pos = int(input("Enter the position (1 to N) to add the task: "))
            todo_list.add_custom(pos, task)
        elif choice == "7":
            pos = int(input("Enter the position (1 to N) to remove the task: "))
            todo_list.remove_custom(pos)
        elif choice == "8":
            todo_list.display()
        elif choice == "9":
            break
        else:
            print("Invalid choice. Please select a valid option (1-9).")

if __name__ == "__main__":
    main()
```
